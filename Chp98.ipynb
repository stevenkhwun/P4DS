# Random Walks example in p. 118 of the book

```python
import matplotlib.pyplot as plt
```

```python
import random
position = 0
walk =[position]
nsteps = 1000
for _ in range(nsteps):
  step = 1 if random.randint(0, 1) else -1
  position += step
  walk.append(position)
  
```

```python
plt.plot(walk[:100])
```
