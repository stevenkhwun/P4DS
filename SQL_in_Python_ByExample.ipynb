# SQL in Python - By Example

This notebook is based on the book "PROC SQL by Example - Using SQL within SAS" by Howard Schreier.

````python
import pandas as pd
import numpy as np
````

````python
# Create the dataset
preteen = pd.DataFrame(
    {"FName": ["James", "Jane", "John", "Joyce", "Louise", "Robert", "Thomas"],
     "Sex": ["M", "F", "M", "F", "F", "M", "M"],
     "Age": [12, 12, 12, 11, 12, 12, 11],
     "Height": [57.3, 59.8, 59.0, 51.3, 56.3, 64.8, 57.5],
     "Weight": [83.0, 84.5, 99.5, 50.5, 77.0, 128.0, 85.0]}
)
preteen
````

## SELECT Statement

```SAS
# SAS SELECT statement
SELECT  *
FROM    preteen;
```

````python
preteen
````

### A more Selective SELECT

```SAS
# More selective SELECT
SELECT  fname, age
FROM    preteen;
```

````python
preteen[["FName", "Age"]]
````

### New columns

```SAS
# Create new columns
SELECT  *,
        weight / height AS Ratio FORMAT=5.2
FROM    preteen;
```

````python
preteen.assign(Ratio = preteen["Weight"]/preteen["Height"])
````
